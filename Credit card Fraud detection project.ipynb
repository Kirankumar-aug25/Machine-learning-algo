```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading dataset using pandas
crd = pd.read_csv(r"D:\bin\creditcard.csv\creditcard.csv")
```

```python
#seperation valid vs fraud for data analysis
valid =crd [crd.Class==0]
fraud = crd [crd.Class==1]
```

```python
#dealing with unbalance data 
#we are to take sample data
# in this we are going to build sample data from original data which is similar distribution of valid and fraud trancstions
```

```python
#random sampling
valid_sample = valid.sample(492)
```

```python
#concatinate
new_crd = pd.concat([valid_sample,fraud],axis =0)
```

```python
new_crd.head(5)
```

```python
new_crd.tail(5)
```

```python
new_crd['Class'].value_counts()
```

```python
new_crd.groupby('Class').mean()
```

```python
#split the data into features and target
X = new_crd.drop(columns= 'Class', axis = 1)
y = new_crd['Class']
```

```python
print(X)
```

```python
print(y)
```

```python
#to split the data into training data and testing data 
#random_state is to get original when u give any number
#stratify?
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size= 0.2,stratify = y,random_state = 2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# TRANING OUR MODEL USING LODGISTIC REGRESSION 
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# Model evaluation based on accuracy 
X_train_predict =model.predict(X_train) 

traning_data_accuracy = accuracy_score(X_train_predict,y_train)
```

```python
print(traning_data_accuracy)
```

```python
# accuracy based on s_test
X_test_predict =model.predict(X_test) 

test_data_accuracy = accuracy_score(X_test_predict,y_test)
```

```python
print(test_data_accuracy)
```
